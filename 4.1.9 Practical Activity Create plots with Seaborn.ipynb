## 4.1.9 Create plots with Seaborn

```python
# Import packages and read data files
```

```python
import matplotlib as plt
import seaborn as sns
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ott = pd.read_excel('ott.xlsx')
```

```python
print(movies.shape)
print(movies.dtypes)
print(movies.columns)
```

```python
movies.head()
```

```python
print(ott.shape)
print(ott.dtypes)
print(ott.columns)
```

```python
ott.head()
```

### Merge the dataframes

```python
# Left Merge with movies being left
# Merge on ID
```

```python
movies_ott = pd.merge(movies, ott, how='left', on='ID')
```

```python
movies_ott.columns
```

```python
movies_ott.head()
```

```python
# Create a countplot to determin which age group has downloaded the most movies
```

```python
sns.countplot(x='Age', hue='Netflix', data=movies_ott)
```

Q2 A The 18+ age group has streamed the most movies

Q2 B Netflix does not have the majority share of each streaming service by any of the age groups.

```python
sns.countplot(x='Age', data=movies_ott)
```

```python
# Create a histogram based on IMDb. 
sns.histplot(data=movies_ott, x='IMDb', binwidth=1)
```

Q3 A Most ratings are between 5 and 7. Distribution is left skewed.

Q3 B Data outliers are scores of 8 to 10.

```python
# Create a scatterplot to show correclation between ratings for Rotten Tomatoes and IMDb
```

```python
sns.scatterplot(x='Rotten Tomatoes', y='IMDb', data=movies_ott)
```

```python
sns.scatterplot(x='Rotten Tomatoes', y='IMDb', legend='full', data=movies_ott)
```

Q4 A Data suggests that higher IMDb ratings have a lower corresponding Rotten Tomatoes rating

Q4 B Moderate, negative correlation
